## Question 1  
Récupérer les 5 projets ayant reçu le plus de promesse de dons.

```python
cur2 = collection.find({}, {"name":1,"usd pledged":1}).sort([("usd pledged", -1)])
next(cur2)
lq1 = list(cur2)
print(lq1[:5])
```

```python
# autre solution
cur = collection.find().sort([("backers",-1)]).limit(5)
list(cur)
```

## Question 3
